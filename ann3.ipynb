```python
# 3. Write a Python Program using Perceptron Neural Network to recognise
# even and odd numbers. Given numbers are in ASCII form 0 to 9
#American Standard Code for Information Interchange

# ASCII of '0' = 48  -> 110000
import numpy as np

#step activation function
sfunc=lambda x: 1 if x>=0 else 0
```

```python
training_data=[
    {'input': [1,1,0,0,0,0], 'label': 1},  # 48 -> 0 -> Even
    {'input': [1,1,0,0,0,1], 'label': 0},  # 49 -> 1 -> Odd
    {'input': [1,1,0,0,1,0], 'label': 1},  # 50 -> 2 -> Even
    {'input': [1,1,0,0,1,1], 'label': 0},  # 51 -> 3 -> Odd
    {'input': [1,1,0,1,0,0], 'label': 1},  # 52 -> 4 -> Even
    {'input': [1,1,0,1,0,1], 'label': 0},  # 53 -> 5 -> Odd
    {'input': [1,1,0,1,1,0], 'label': 1},  # 54 -> 6 -> Even
    {'input': [1,1,0,1,1,1], 'label': 0},  # 55 -> 7 -> Odd
    {'input': [1,1,1,0,0,0], 'label': 1},  # 56 -> 8 -> Even
    {'input': [1,1,1,0,0,1], 'label': 0}   # 57 -> 9 -> Odd
]
weights = np.array([0, 0, 0, 0, 0, 1])

#training
for data in training_data:
    x=np.array(data['input'])
    label=data['label']
    output=sfunc(np.dot(x,weights))
    error=label-output
    weights=weights + error*x

char=input("enter a digit (0-9):")
if char not in "0123456789":
    print("invalid input")
else:
    #convert to ascii
    ascii_value=ord(char)#convert char to ascii value the ascii to binary
    test_input=np.array([int(x) for x in format(ascii_value,'06b')])
    result=sfunc(np.dot(test_input,weights))
    if result==1:
        print( char ,"is even")
    else:
        print(char,"is odd")
print("learned weights",weights)
```
