```python
import numpy as np
x=np.array([[1,0,1]]) #1*3 shape
y=np.array([[1]])
def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoid_dervetive(a):
    return a*(1-a)

#initilizing weights and bias
w=np.random.rand(3,1)  # shape 3*1
b=0
lr=0.5

#training loop
for epochs in range(10):
    #forward propogation
    z=(np.dot(x,w))+b
    output=sigmoid(z)
    # loss calculation mse
    loss=0.5*(y-output)**2

    #back propogation using gd
    #dl/dw=dl/da*da/dz*dz/dw
    #dl/dw=(output-y)*deravetive sigmoid(output)*x
    d_loss=(output-y)
    d_output=d_loss*sigmoid_dervetive(output)


    w -= lr*np.dot(x.T,d_output)
    b -= lr*d_output.item()


    print(f"epoch {epochs +1}:output={output},loss={loss}")

print("final output")
print(output)
print("final weight")
print(w)
print("final bias")
print(b) 
```
