```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,100)
```

```python
#linear activation function
def linear(x):
    return x
y=linear(x)
plt.plot(x,y)
plt.title("linear activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
```

```python
def sigmoid(x):
    
    return 1/(1+np.exp(-x))
y=sigmoid(x)
plt.plot(x,y)
plt.title("sigmoid activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
```

```python
def tanh_x(x):
    
    return np.tanh(x)
y=tanh_x(x)
plt.plot(x,y)
plt.title("tanh activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
```

```python
def relu(x):
    
    return np.maximum(0,x)
y=relu(x)
plt.plot(x,y)
plt.title("relu activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
```

```python
def softmax(x):
    
    return np.exp(x)/np.sum(np.exp(x))
y=softmax(x)
plt.plot(x,y)
plt.title("softmax activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
```

```python
def step(x):
    
    return np.where(x >= 0, 1, 0)
y=step(x)
plt.plot(x,y)
plt.title("step activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
```

```python
def bipolar(x):
    
    return np.where(x >= 0, 1, -1)
y=bipolar(x)
plt.plot(x,y)
plt.title("bipolar activation function")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
```
