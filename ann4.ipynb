```python
# 4. With a suitable example demonstrate the perceptron learning law
# with its decision regions using python.
#     Give the output in graphical form. is this correct
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defining and gate
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])

w=np.zeros(x.shape[1])
b=0
lr=0.3
step=lambda x: 1 if x>=0 else 0
for epochs in range(10):
    for i in range(x.shape[0]):
        y_pred=step(np.dot(x[i],w)+b)
        
        if y[i] != y_pred:
            
            e=y[i]-y_pred
            w=w+e*lr*x[i]
            b=b+e*lr
        


# Final Parameters
print("Final Weights:", w)
print("Final Bias:", b)


# Plot Decision Regions
# xx, yy = np.meshgrid(
#     np.arange(-1, 2, 0.01),
#     np.arange(-1, 2, 0.01)
# )

# Z = np.array([
#     step(np.dot([x1, x2], w) + b)
#     for x1, x2 in zip(xx.ravel(), yy.ravel())
# ]).reshape(xx.shape)

# plt.contourf(xx, yy, Z, alpha=0.4)

# plt.scatter(
#     x[:,0],
#     x[:,1],
#     c=y,
#     edgecolors='k',
#     s=200
# )
xx,yy=np.meshgrid(
    np.arange(-1,2,0.01),
    np.arange(-1,2,0.01)
)

z=np.array([step(np.dot([x1,x2],w)+b) for x1,x2 in zip(xx.ravel(),yy.ravel())]).reshape(xx.shape)
plt.contourf(xx,yy,z,alpha=0.4)
plt.scatter(x[:,0],x[:,1])

plt.xlabel("X1")
plt.ylabel("X2")

plt.title("Perceptron Learning Law with Decision Regions")

plt.grid(True)

plt.show()
```
