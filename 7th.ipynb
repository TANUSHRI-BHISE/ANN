```python
import numpy as np
# Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(a):
    return a * (1 - a)
# XOR Input & Output
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

Y = np.array([[0],[1],[1],[0]])

# Initialize Parameters
np.random.seed(1)

w1 = np.random.rand(2,2)   # input → hidden
b1 = np.zeros((1,2))

w2 = np.random.rand(2,1)   # hidden → output
b2 = np.zeros((1,1))

lr = 0.5
# Training
for epoch in range(10000):

    # 🔹 Forward Propagation
    z1 = np.dot(X, w1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, w2) + b2
    output = sigmoid(z2)

    # 🔹 Loss (MSE)
    loss = np.mean((Y - output)**2)

    # 🔹 Backpropagation (CHAIN RULE)
    d_output = (output - Y) * sigmoid_derivative(output)

    d_hidden = np.dot(d_output, w2.T) * sigmoid_derivative(a1)

    # 🔹 Gradient Descent Update
    w2 -= lr * np.dot(a1.T, d_output)
    b2 -= lr * np.sum(d_output, axis=0)

    w1 -= lr * np.dot(X.T, d_hidden)
    b1 -= lr * np.sum(d_hidden, axis=0)

    # Print loss occasionally
    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Final Output
print("\nFinal Output:")
print(output.round())
```
